## 2. Variables and Basic Data Types in Python

Python supports several basic data types that are very useful in data science:

- **int** â€“ integers (e.g. 1, 20, -5)  
- **float** â€“ decimal numbers (e.g. 3.14, 0.5, -2.0)  
- **str** â€“ strings (text) (e.g. `"OOU"`, `"Data Science"`)  
- **bool** â€“ boolean values (`True` or `False`)  

We assign values to **variables** using the `=` sign.


````
# Create an integer variable representing a student's age
age = 20  # 'age' stores the whole number 20

# Create a float variable representing a student's exam score
score = 72.5  # 'score' stores a decimal number (float)

# Create a string variable representing a student's name
name = "Amina"  # 'name' stores a sequence of characters (text)

# Create a boolean variable representing whether the student passed
passed = True  # 'passed' is True if the student passed, False otherwise

# Print the variables and their types
print("Age:", age, "| Type:", type(age))
print("Score:", score, "| Type:", type(score))
print("Name:", name, "| Type:", type(name))
print("Passed:", passed, "| Type:", type(passed))
````

## 3. Basic Arithmetic and Comparisons

Python can be used as a simple calculator for arithmetic operations and for comparing values.

Common arithmetic operators:

- `+` (addition), `-` (subtraction), `*` (multiplication), `/` (division), `//` (integer division), `%` (modulo), `**` (power).

Comparison operators (return `True` or `False`):

- `==`, `!=`, `>`, `<`, `>=`, `<=`.


````
# Example: basic arithmetic with numbers
a = 10        # First number
b = 3         # Second number

sum_ab = a + b        # Addition
diff_ab = a - b       # Subtraction
prod_ab = a * b       # Multiplication
quot_ab = a / b       # Division (float result)
int_div = a // b      # Integer division (no remainder)
mod_ab = a % b        # Remainder after division
power_ab = a ** b     # a raised to the power b

print("a + b =", sum_ab)
print("a - b =", diff_ab)
print("a * b =", prod_ab)
print("a / b =", quot_ab)
print("a // b =", int_div)
print("a % b =", mod_ab)
print("a ** b =", power_ab)
````

````
# Example: comparison operators
x = 75       # A sample exam score
y = 50       # Pass mark

print("x > y:", x > y)      # Is x greater than y?
print("x == y:", x == y)    # Is x equal to y?
print("x >= y:", x >= y)    # Is x greater than or equal to y?
print("x < y:", x < y)      # Is x less than y?
````

## 4. Working with Lists

A **list** is an ordered collection of items. Lists can store numbers, strings, or even mixtures of types.

Lists are defined using **square brackets**: `[ ... ]`.


````
# Create a list of exam scores
scores = [72, 65, 88, 90, 54]  # A list containing five integers

# Print the entire list
print("Scores list:", scores)

# Access individual elements by index (starting at 0)
print("First score:", scores[0])   # Element at index 0
print("Third score:", scores[2])   # Element at index 2

# Add a new score to the end of the list
scores.append(79)  # Append 79 to the list

# Print updated list
print("Updated scores list:", scores)

# Compute simple statistics using built-in functions
print("Number of scores:", len(scores))   # len() gives the list length
print("Maximum score:", max(scores))      # max() gives the largest value
print("Minimum score:", min(scores))      # min() gives the smallest value
print("Average score:", sum(scores) / len(scores))  # Average = sum / count
````

## 5. Dictionaries (Keyâ€“Value Pairs)

A **dictionary** stores data as **keyâ€“value pairs**.

Example:


You access values using keys like `student["name"]`.


````
# Create a dictionary to store a student's information
student = {
    "name": "Amina",                # Key 'name' maps to the string "Amina"
    "age": 20,                      # Key 'age' maps to the integer 20
    "department": "Computer Science",  # Key 'department' maps to a string
    "passed": True                  # Key 'passed' maps to a boolean
}

# Access values by their keys
print("Student name:", student["name"])
print("Student age:", student["age"])
print("Department:", student["department"])
print("Passed?:", student["passed"])

# Add a new keyâ€“value pair for 'level'
student["level"] = 200  # Add a new entry with key 'level' and value 200

print("Updated student dictionary:", student)
````

## 6. Decision-Making with `if`, `elif`, and `else`

We use **if-statements** to run certain blocks of code only when a condition is `True`.

````
# Example: grade classification based on score
score = 72  # You can change this value to test different branches

# Use if/elif/else to assign a grade based on score
if score >= 70:
    grade = "A"       # If score is 70 or higher, grade is A
elif score >= 60:
    grade = "B"       # If score is between 60 and 69, grade is B
elif score >= 50:
    grade = "C"       # If score is between 50 and 59, grade is C
else:
    grade = "F"       # Any score below 50 gets an F

print("Score:", score)
print("Grade:", grade)
````

## 7. Loops: Repeating Actions with `for`

Loops allow us to repeat operations over lists or ranges without writing the same line many times.

````
# Example: print each score in a list and check if it is a pass or fail
scores = [72, 65, 88, 90, 54]  # List of exam scores
pass_mark = 50                 # Pass mark threshold

# Loop over each score in the list
for s in scores:
    # For each score, check if it meets the pass mark
    if s >= pass_mark:
        status = "PASS"  # Status is PASS if score >= pass_mark
    else:
        status = "FAIL"  # Status is FAIL otherwise

    # Print out the score and its status
    print(f"Score: {s} â†’ {status}")
````

## 8. Writing Your Own Functions

A **function** is a reusable block of code that performs a specific task.  

We define functions using the `def` keyword.


````
# Define a function to compute the average of a list of numbers
def compute_average(values):
    """Return the average (mean) of a list of numerical values."""
    total = sum(values)         # Compute the sum of all values in the list
    count = len(values)         # Count how many values are in the list
    average = total / count     # Divide sum by count to get the mean
    return average              # Return the computed average

# Use the function on a list of scores
scores = [72, 65, 88, 90, 54]

avg_score = compute_average(scores)  # Call the function and store the result

print("Scores:", scores)
print("Average score:", avg_score)
````

## 9. Basic Data Operations with `pandas`

`pandas` is a core library for data science in Python.  
It provides the **DataFrame**, which is like an improved, programmable spreadsheet.

Here we will create a small DataFrame and perform basic operations.


````
# Import pandas (if not already imported above)
import pandas as pd  # Alias 'pd' is a common convention

# Create a simple DataFrame from a dictionary
data = {
    "Name": ["Amina", "Bola", "Chidi", "Damilola"],  # List of student names
    "Score": [72, 65, 88, 54],                      # Corresponding scores
    "Department": ["CS", "CS", "IT", "CS"]          # Departments
}

students_df = pd.DataFrame(data)  # Convert the dictionary into a DataFrame

# Display the entire DataFrame
students_df
````

````
# View basic information about the DataFrame
print("Shape (rows, columns):", students_df.shape)  # Number of rows and columns
print("\nInfo:")
print(students_df.info())  # Data types and non-null counts
````

````
# Select a single column
scores_series = students_df["Score"]  # Series of scores
print("Scores column:")
print(scores_series)

# Compute average score using pandas
print("\nAverage score (pandas):", students_df["Score"].mean())
````

````
# Filter rows where Score is >= 70
passed_df = students_df[students_df["Score"] >= 70]  # Select only rows with Score >= 70

print("Students who scored 70 and above:")
print(passed_df)
````
